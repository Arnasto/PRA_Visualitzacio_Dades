```python
df['DATA'] = pd.to_datetime(df['DATA'])
df['Year'] = df['DATA'].dt.year

# Reorganitzem les columnes
df_melted = df.sample(100000, random_state=42).melt(
    id_vars=['NOM ESTACIO', 'DATA', 'Year', 'MAGNITUD', 'CONTAMINANT', 'UNITATS', 'TIPUS ESTACIO', 'AREA URBANA', 'CODI INE', 'MUNICIPI', 'CODI COMARCA', 'NOM COMARCA', 'LATITUD', 'LONGITUD'],
    value_vars=[f'{i:02d}h' for i in range(1, 25)],
    var_name='Hour',
    value_name='Value'
)

df_melted['Hour'] = df_melted['Hour'].str.slice(0, 2).astype(int)

# Promig per hores de cada any
hourly_avg_df = df_melted.groupby(['Year', 'MUNICIPI', 'Hour', 'CONTAMINANT'])['Value'].mean().reset_index()

# Promig anual per capitals de provincia
main_cities = ['Barcelona', 'Tarragona', 'Girona', 'Lleida']
yearly_avg_df = df_melted[df_melted['MUNICIPI'].isin(main_cities)].groupby(['Year', 'MUNICIPI', 'CONTAMINANT'])['Value'].mean().reset_index()

print("Hourly Average DataFrame:")
print(hourly_avg_df.head())

print("\nYearly Average DataFrame for Main Cities:")
print(yearly_avg_df.head())
```

```python
import pandas as pd
import plotly.graph_objects as go

# Sample
df_sampled = df.sample(100000, random_state=42)
df_sampled['DATA'] = pd.to_datetime(df_sampled['DATA'], format="%d/%m/%Y")

# Fixem any
fixed_year = 2024

# Filtrem dades per NO2
df_no2 = df_sampled[df_sampled['CONTAMINANT'] == 'NO2']

# Calcul dels promitjos diaris en un any
def get_daily_avg_data(year):
    filtered_df = df_no2[df_no2['DATA'].dt.year == year]
    filtered_df = filtered_df.melt(
        id_vars=['NOM ESTACIO', 'LATITUD', 'LONGITUD'],
        value_vars=[f'{i:02d}h' for i in range(1, 25)],
        var_name='Hour',
        value_name='Value'
    )
    filtered_df['Value'] = filtered_df['Value'].astype(float)
    daily_avg_df = filtered_df.groupby(['LATITUD', 'LONGITUD'])['Value'].mean().reset_index()
    return daily_avg_df

daily_avg_df = get_daily_avg_data(fixed_year)

# Plot mapa de densitat
fig = go.Figure(go.Densitymapbox(lat=daily_avg_df['LATITUD'], lon=daily_avg_df['LONGITUD'], z=daily_avg_df['Value'],
                                 radius=10, colorscale='Viridis', zmax=daily_avg_df['Value'].max(), zmin=daily_avg_df['Value'].min(),
                                 showscale=True))

fig.update_layout(mapbox_style="open-street-map", mapbox_center_lon=2.1734, mapbox_center_lat=41.3851,
                  mapbox_zoom=6, title=f'NO2 Pollution Density Map in {fixed_year}')

fig.update_layout(margin={"r":0,"t":30,"l":0,"b":0})
fig.show()

##fig.write_html('hourly_avg_no2_heatmap.html')
```

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

# Sample
df_sampled = df.sample(100000, random_state=42)

# Convertim
df_sampled['DATA'] = pd.to_datetime(df_sampled['DATA'], format="%d/%m/%Y")

#Filtrem Barcelona i NO2
df_barcelona_no2 = df_sampled[(df_sampled['MUNICIPI'] == 'Barcelona') & (df_sampled['CONTAMINANT'] == 'NO2')]

# Extraiem dades
df_barcelona_no2['Year'] = df_barcelona_no2['DATA'].dt.year
df_barcelona_no2['Hour'] = df_barcelona_no2['DATA'].dt.hour

# Agrupem i calculem nivells de NO2
hourly_avg_df = df_barcelona_no2.groupby(['Year', 'Hour'])[['01h', '02h', '03h', '04h', '05h', '06h',
                                                           '07h', '08h', '09h', '10h', '11h', '12h',
                                                           '13h', '14h', '15h', '16h', '17h', '18h',
                                                           '19h', '20h', '21h', '22h', '23h', '24h']].mean().reset_index()

hourly_avg_df = hourly_avg_df.melt(id_vars=['Year', 'Hour'], var_name='Hourly', value_name='Average Concentration')

# Definim escala de colors
color_scale = px.colors.sequential.Plasma
num_years = len(hourly_avg_df['Year'].unique())
colors = [color_scale[i * (len(color_scale) - 1) // (num_years - 1)] for i in range(num_years)]

fig = go.Figure()

# Afegim linies per cada any
for i, year in enumerate(hourly_avg_df['Year'].unique()):
    df_year = hourly_avg_df[hourly_avg_df['Year'] == year]
    fig.add_trace(go.Scatter(x=df_year['Hourly'], y=df_year['Average Concentration'],
                             mode='lines',
                             name=str(year),
                             line=dict(color=colors[i], width=2),
                             hovertemplate='Year: ' + str(year) + '<br>' +
                                           'Hour: %{x}<br>' +
                                           'Avg Concentration: %{y}<extra></extra>'
                             ))

# Afegim linies adicionals
fig.add_trace(go.Scatter(x=hourly_avg_df['Hourly'], y=[40] * len(hourly_avg_df),
                         mode='lines',
                         line=dict(color='red', width=1, dash='dash'),
                         name='European Limit for NO2 Concentration (40)',
                         hoverinfo='text',
                         hovertext='European Limit for NO2 Concentration: 40'
                         ))

fig.add_trace(go.Scatter(x=hourly_avg_df['Hourly'], y=[20] * len(hourly_avg_df),
                         mode='lines',
                         line=dict(color='blue', width=1, dash='dash'),
                         name='Objective for NO2 Concentration (20)',
                         hoverinfo='text',
                         hovertext='Objective for NO2 Concentration: 20'
                         ))

# Update layout
fig.update_layout(title='Hourly Average NO2 Concentrations Over Hours for Each Year in Barcelona',
                  xaxis_title='Hour of the Day',
                  yaxis_title='Average Concentration',
                  legend_title='Year')

fig.show()

##fig.write_html('hourly_avg_no2_barcelona.html')
```
